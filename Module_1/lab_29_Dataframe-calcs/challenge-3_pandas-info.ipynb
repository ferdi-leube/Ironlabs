```python
# your code here
customerspending = orders.groupby(['CustomerID','Country'], as_index=False).agg({'amount_spent':'sum'})
customerspending

c = customerspending.quantile(0.95)
c
VIP = customerspending[customerspending['amount_spent']>c[1]]
VIP
```

```python
d = customerspending.quantile(0.75)
pref_customers1 = customerspending[customerspending['amount_spent']<c[1]]
pref_customers = pref_customers1[pref_customers1['amount_spent']>d[1]]
pref_customers
```

```python
## other way of retrieving pref or VIP customers, through cut

spent_per_customer = orders.groupby(['CustomerID', 'Country'], as_index=False).agg({'amount_spent':'sum'})
spent_per_customer
```

```python
agg_min=spent_per_customer['amount_spent'].min()
agg_max=spent_per_customer['amount_spent'].max()
```

```python
q_095 = spent_per_customer['amount_spent'].quantile(0.95)
q_075 = spent_per_customer['amount_spent'].quantile(0.75)

cut_label = [0,'prefcustom','VIP']

cutoffs = [agg_min, q_075, q_095, agg_max]
spent_per_customer["bins"] = pd.cut(spent_per_customer['amount_spent'], cutoffs, labels=cut_label)
spent_per_customer
```
