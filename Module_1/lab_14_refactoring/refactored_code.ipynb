```python
def input_check(x):
    flag1 = False
    user1=str(x) #turning input into string
    x = x.lower() #lower case the word
    while flag1 == False:
        if x.strip().isdigit() == False:
            return(str(x))
            flag1 = True
        else:
            x = input("Error: Please enter a !WORD/LETTERS!: ")
```

```python
#define global variables
life = 12 
lst_of_input_letters = []
lst_og_letters = []
```

```python
#user 1 inputs word and check if inputted string is number or word

user1 = input("Player 1: Please enter your word: ")
user1 = input_check(user1) 

#create maximum length of word  

flag1 = False
while flag1 == False:
    if len(user1) <= 26:
        break
    else:
        user1 = input("Error: Please enter a word with maximum length of 26 letters: ")
        flag1 = True
```

```python
# now set the base for the game to work --> lists containing number of positions, dictionaries containing positions of unique letters etc.

# create a list of user1 input

lst_og_letters=list(user1)

#create an empty dictionary

d=dict()

#turning the list of letters into a set --> only unique values saved here, in case of double letters, only one letter present

cl = set(lst_og_letters)

#creating a list, which consists out of the postions of the letters in the word

answers_position_list=list(range(0, len(lst_og_letters)))

#for the unique values of the word(in set) we know want to find the position of this value in the letter, and add this relationship in form of a key value pair (character:position) into the dictionary

for c in cl:
    lst =[]
    for pos,char in enumerate(user1):
        if(char == c):
            lst.append(pos)
            d[c]=lst
```

```python
#a while loop is set up to make the game continue, ask user 2 for new letters until the game is either won or lost

while life > 0 and lst_og_letters != answers_position_list:
    
        #input of user 2 and check if word or number
        user2 = input("Player 2: Please enter a letter: ")
        user2 = input_check(user2)
        
        #now we make conditional checks to see if input letter is in word, not in word, or has already been used and apply the consequences according to the rules of the game
        if user2 in lst_og_letters and user2 not in lst_of_input_letters:
            lst_of_input_letters.append(user2)
            
            for i in d[user2]: #we do this in case there are multiple positions for one letter, to assign letter to all positions saved in list of letter key of d
                answers_position_list[i]=user2 #then replace the position/positions with letter y 
                
            # I want to print out the contents next to each other, not in a list
            output = None
            for items in answers_position_list:
                items=str(items)
                output +=items + ' '
            print(output)
            
        elif user2 not in lst_og_letters and user2 not in lst_of_input_letters:
            lst_of_input_letters.append(user2)
            life_function(life)
            print(user2, "is .... not in the word :)")
            life -= 1
            print("You have", life, "lives left!!")
            
        elif user2 in lst_of_input_letters:
            print("letter has already been used")
            
#last step: we reach this point when while loop has broken, that means either one of the players has now won, results are then printed       

if life == 0:
    life_function(0)
    print("PLAYER 1 WINS!!")
elif answers_position_list == lst_og_letters:
    print("PLAYER 2 WINS!!")   
```
