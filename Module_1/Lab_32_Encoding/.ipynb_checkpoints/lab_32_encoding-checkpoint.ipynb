```python
import pandas as pd
import numpy as npPr
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({'TransactionID':[1,2,3,4,5],
                   'ClientID':[231,765,453,231,892],
                   'Profession':['Self-employed','students','Horeca','self-employed','finance'],
                   'Bank_dep':['009','005','007','009','003'],'Risk':['high','high','medium','high','low'],
                   'Number of credits':[2,3,5,2,3]})
```

```python
categoricaldata = df.select_dtypes('object')
del categoricaldata['Bank_dep']
```

```python
categoricaldata
```

```python
categoricaldata['Profession'] = categoricaldata['Profession'].str.lower()
```

```python
# dummy encoder
pd.get_dummies(categoricaldata)
```

```python
# label encoder
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(categoricaldata['Profession'])
le.transform(categoricaldata['Profession'])
categoricaldata1 = categoricaldata
categoricaldata1['encoded'] = le.transform(categoricaldata['Profession'])

```

```python
from sklearn.preprocessing import LabelEncoder
categoricaldata['label_encoder'] = LabelEncoder().fit_transform(categoricaldata.Risk)
categoricaldata
```

```python
# ordinal
num_dict = {'self:employed':0,
            'stusents':1,
            'horeca':2,
            'finance':3}

categoricaldata2 = categoricaldata
categoricaldata2['num_ordinal'] = categoricaldata2.Profession.map(num_dict)
```

```python
## https://towardsdatascience.com/all-about-categorical-variable-encoding-305f3361fd02#:~:text=In%20backward%20difference%20coding%2C%20the,coding%20system%20for%20categorical%20features.
# https://contrib.scikit-learn.org/category_encoders/basen.html
# link to good website describing different encoding methods
```

```python
# Backward Difference Coding
categoricaldata4 = categoricaldata
import category_encoders as ce
encoder = ce.BackwardDifferenceEncoder(cols=['Risk'])
df_bd = encoder.fit_transform(categoricaldata4)

df_bd.head()
```

```python
# Helmet encoding
import category_encoders as ce
encoder = ce.helmert.HelmertEncoder(cols=['Risk'])
encoder.fit_transform(df['Risk'], y=None)
```

```python
#  binary encoding
categoricaldata3 = categoricaldata
import category_encoders as ce
encoder = ce.BinaryEncoder(cols='Profession')
dfbin = encoder.fit_transform(categoricaldata3['Profession'])
categoricaldata3 = pd.concat([categoricaldata3,dfbin],axis=1)
categoricaldata3
```

```python
# BaseN encoder
encoder = ce.basen.BaseNEncoder(cols=['Risk'])
encoder.fit_transform(df['Risk'], y=None)
```

```python
# cat boost encoder

encoder = ce.cat_boost.CatBoostEncoder(cols=['Risk'])
encoder.fit_transform(df['Risk'], y=[1,2,3,4,5])
```
