```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv(r'C:\Users\leube\Downloads\example_data_cleaning.xlsx - Sheet1.csv')
```

```python
del data['Department']
```

```python
data
```

```python
data['Profession'] = data['Profession'].str.lower()
data
```

```python
data.loc[data['Profession'].str.startswith('etu'), 'Profession'] = 'student'
```

```python
data['Profession'] = data['Profession'].ffill()
data['BirthYear'] = data['BirthYear'].ffill()
data
```

```python
a = data['Amount'].max()
b = data['Amount'].median()
c= data['Amount'].min()
data.replace([a,c],b)
```

```python
# take out clients older than born in 1930
data = data[data.BirthYear > 1930]
data
```

```python
# check if every client only has one birth year corresponded to them
l = data['ClientID']
for i in l:
    print(i)
    print(data[data['ClientID'] == i])
    print('\n')
```

```python
def change_year(clientID, changetoyear):
    data.loc[data['ClientID']==clientID, 'BirthYear'] = changetoyear
```

```python
change_year(34988,1999)
change_year(34989,1958)
change_year(34987,1967)
change_year(35008,1967)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['Profession_label_encoded'] = LabelEncoder().fit_transform(data.Profession)
```

```python
data
```

```python
def encode_risk(risk,value):
    risk = str(risk)
    value = int(value)
    data.loc[data['Risk']== f'{risk}', 'Risk_importance_encoded'] = value
```

```python
encode_risk('Low',0)
encode_risk('Medium',1)
encode_risk('High',2)
```
