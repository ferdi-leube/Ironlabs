For the second part of this lab, we will use the vehicles.csv data set. You can find a copy of the dataset in the git hub folder. This dataset includes variables related to vehicle characteristics, including the model, make, and energy efficiency standards, as well as each car's CO2 emissions. As discussed in class the goal of this exercise is to predict vehicles' CO2 emissions based on several independent variables. 

```python
# Import any libraries you may need & the data
data = pd.read_csv(r"C:\Users\leube\Ironhack\cloned_labs\module_2\Lab_16_Linear-Regression\vehicles.csv")
v = pd.DataFrame(data)
```

```python
v.head()
```

Let's use the following variables for our analysis: Year, Cylinders, Fuel Barrels/Year, Combined MPG, and Fuel Cost/Year. We will use 'CO2 Emission Grams/Mile' as our outcome variable. 

Calculate the correlations between each of these variables and the outcome. Which variable do you think will be the most important in determining CO2 emissions? Which provides the least amount of helpful information for determining CO2 emissions? Add your responses as comments after your code.

```python
# Your response here. 
# create correlation matrix of specified variables
variables = v[['Year', 'Cylinders', 'Fuel Barrels/Year', 'Combined MPG', 'Fuel Cost/Year', 'CO2 Emission Grams/Mile']]
```

```python
variables.corr()
```

positive correlations --> Fuel Barrels per year has the highest correlation, because we are looking for which factor is responsible for Co2 emission increase, we will not need to look into combined MPG because it has a strong negative correlation

Build a regression model for this data. What type of regression are you using? Add your responses as comments after your code.

```python
# Your response here


X = sm.add_constant(v[['Year', 'Cylinders', 'Fuel Barrels/Year', 'Combined MPG', 'Fuel Cost/Year']])
Y = v['CO2 Emission Grams/Mile']
model = sm.OLS(Y,X)
results = model.fit()
predictions = results.predict(X)
```

```python
predictions
```

```python
residuals = [Y[i] - predictions[i] for i in range(len(Y))]
```

```python
residuals
```

```python
import numpy as np
RMSE = np.sqrt(np.mean([residual**2 for residual in residuals]))
```

```python
RMSE
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(mean_squared_error(Y, predictions))
print(r2_score(Y, predictions))
print(np.sqrt(mean_squared_error(Y, predictions)))
```

Print your regression summary, and interpret the results. What are the most important varibles in your model and why? What can conclusions can you draw from your model and how confident in these conclusions are you? Add your responses as comments after your code.

```python
import statsmodels.api as sm

print_model = results.summary()
print(print_model)
```

high r squared value --> very high, refelcts about 99% of the variation of the data, consider the predictions as accurate with high confidence
