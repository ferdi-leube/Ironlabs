#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
fig, [p1,p2] =plt.subplots(1,2)

p1.plot(x, y, label='1')

p1.set_xlabel('x')
p1.set_ylabel('y')

p1.set_title('p1')
p1.legend()
# 

p2.plot(x, z, label='2')

p2.set_xlabel('x')
p2.set_ylabel('y')

p2.set_title('p2')
p2.legend()
plt.style.use('ggplot')
plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here
fig, [p1,p2] =plt.subplots(1,2)

fig.set_figheight(7) ##
fig.set_figwidth(10) ##

p1.plot(x, y, label='1')

p1.set_xlabel('x')
p1.set_ylabel('y')

p1.set_title('p1')
p1.legend()
# 

p2.plot(x, z, label='2')

p2.set_xlabel('x',size=20)
p2.set_ylabel('y',size=20)

p2.set_title('p2', fontweight= 'bold', size=20)
p2.legend()
plt.style.use('ggplot')
plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
fig, [p1,p2] =plt.subplots(1,2)

fig.set_figheight(7) ##
fig.set_figwidth(10) ##

p1.plot(x, x**2, label='1')

p1.set_xlabel('x')
p1.set_ylabel('y')

p1.set_title('p1')
p1.legend()
# 

p2.plot(x, np.exp(x), label='2')
p2.set_xscale('linear')
p2.set_yscale('log')
p2.set_xlabel('x',size=20)
p2.set_ylabel('y',size=20)

p2.set_title('p2', fontweight= 'bold', size=20)
p2.legend()
plt.style.use('ggplot')
plt.show()
```
