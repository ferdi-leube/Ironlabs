#### Find the minimum, mean and maximum of all numeric columns for each company department.



```python
# your code here
lst = list(employees['Department'].unique())
```

```python
def describe_department(DM):
    filtered = employees[employees['Department']==DM]
    print('\n',DM)
    return print(filtered.describe())
    
for x in lst:
    describe_department(x)
```
